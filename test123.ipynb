```python
#Imorting Libraries

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
url = "https://raw.githubusercontent.com/StephenJudeD/Apple-Stock-Price-Data-Visualization-/main/TimeSeries_AppleStockPrice.csv"
#df = pd.read_csv(url,index_col=7)
df = pd.read_csv(url)

print(df.head(5))
```

```python
#Creating a seperate df to remove volume
df2 = df[['Date','Open', 'High', 'Low','Close', 'Adj Close']]

df2.head()
```

```python
result = df2.dtypes

print("Output:")
print(result)
```

```python
#changing date format to diplay better on plotly

df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d')
```

```python
# apply the dtype attribute
result = df.dtypes

print("Output:")
print(result)
```

```python
# A simple scatter with date and stock high
fig = go.Figure([go.Scatter(x=df['Date'], y=df['High'])])
fig.show()
```

**Figure 1: Interactive Value by Month**

```python
#using df so we have consistancy accross close variables
#https://i.stack.imgur.com/lFZum.png = colours availabloe
fig = px.line(df, x="Date", y=df2.columns,
              hover_data={"Date": "|%B %d, %Y"},
              title='Interactive: Value by Month')
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")

fig['data'][0]['line']['color']='yellowgreen'
fig['data'][0]['line']['width']=2

fig['data'][1]['line']['color']='darkred'
fig['data'][1]['line']['width']=2

fig['data'][2]['line']['color']='red'
fig['data'][2]['line']['width']=2

fig['data'][3]['line']['color']='limegreen'
fig['data'][3]['line']['width']=2

fig['data'][4]['line']['color']='darkgreen'
fig['data'][4]['line']['width']=2



fig.show()
```

```python

fig = px.histogram(df, x="Date", y="Close", histfunc="avg", title="Histogram on Date Axes")
fig.update_traces(xbins_size="M1")

fig.update_xaxes(showgrid=True, ticklabelmode="period", dtick="M1", tickformat="%b\n%Y")
fig.update_layout(bargap=0.1)
fig.add_trace(go.Scatter(mode="markers", x=df["Date"], y=df["Adj Close"], name="Monthly"))



fig.show()
```

```python

fig = px.line(df, x='Date', y='High', title='Time Series of Apple Stock High with Range Slider and Selectors')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```

**Figure 3 & 4: Interactive Value by Month**

```python
fig = px.line(df, x='Date', y='Volume', title='Time Series of Volume with Range Slider and Selectors')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)

fig['data'][0]['line']['color']='black'
fig['data'][0]['line']['width']=2

#https://www.december.com/html/spec/colorrgbadec.html
fig.update_layout({
'plot_bgcolor': 'rgb(192,192,192)',

})

fig.show()
```
